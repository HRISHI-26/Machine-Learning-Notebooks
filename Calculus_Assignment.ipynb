## Gradient Descent Algorithmn

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and its derivative

def f(x):
  return (x - 3)** + 2

def df(x):
  return 2 * (x - 3)


# Gradient Descent Parameters
lr = 0.1
steps = 20
x_vals = [0]


# Run gradient descent
for i in range(steps):
  x_new = x_vals[-1] - lr * df(x_vals[-1])
  x_vals.append(x_new)


# Generate function curve for plotting
x = np.linspace(-1, 6, 100)
y = f(x)


# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x) = (x - 3)^2 + 2", color='blue')
plt.scatter(x_vals, [f(i) for i in x_vals], color='red', label="GD Steps")
plt.plot(x_vals, [f(i) for i in x_vals], linestyle='--', color='orange')
plt.title("Gradient Descent Optimization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

##Integration and Differentiation problems

```python
import sympy as sp
```

```python
# Limits

x, y = sp.symbols('x y')

f = sp.sin(x) / x
lim = sp.limit(f, x, 1)
print("Limit of f(x) where x-> 1: ", lim)
```

```python
# Diffrentiation
f = x**2

df = sp.diff(f, x)
print("Derivative of f(x): ", df)
```

```python
# Partial Diffrentiation

f = x*y

df_x = sp.diff(f, x)
df_y = sp.diff(f, y)
print("Derivative of x: ", df_x)
print("Derivative of y: ", df_y)
```

```python
# Integration

f = 1 / x
ig = sp.integrate(f, x)
print("Integral of f(x): ", ig)
```

```python
# Definite Integration

f = 1 / x
ig  = sp.integrate(f, (x, 1, 2))
print("Integral of f(x): ", ig)
```
